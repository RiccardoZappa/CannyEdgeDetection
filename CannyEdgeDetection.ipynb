```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Load an image (replace 'input_image.jpg' with your image file)
image = cv2.imread('sample_data/sample.jpg', cv2.IMREAD_GRAYSCALE)

# Step 1: Gaussian Smoothing
blurred = cv2.GaussianBlur(image, (3, 3), 0.6)
cv2.imwrite('gaussian_smoothing.png', blurred)

# Step 2: Gradient Computation
sobel_x = cv2.Sobel(blurred, cv2.CV_64F, 1, 0, ksize=3)
sobel_y = cv2.Sobel(blurred, cv2.CV_64F, 0, 1, ksize=3)
gradient_magnitude = np.sqrt(sobel_x**2 + sobel_y**2)
gradient_magnitude = np.uint8(gradient_magnitude)
cv2.imwrite('gradient_computation.png', gradient_magnitude)

# Step 3: Non-Maximum Suppression
edges = cv2.Canny(blurred, 50, 150)
cv2.imwrite('non_max_suppression.png', edges)

# Step 4: Hysteresis Thresholding
# The Canny function in OpenCV already performs hysteresis thresholding
cv2.imwrite('hysteresis_thresholding.png', edges)

# Example of Canny Edge Detection (Final Output)
cv2.imwrite('canny_example.png', edges)

# Display the images (optional)
plt.figure(figsize=(10, 8))
plt.subplot(2, 3, 1), plt.imshow(image, cmap='gray'), plt.title('Original Image')
plt.subplot(2, 3, 2), plt.imshow(blurred, cmap='gray'), plt.title('Gaussian Smoothing')
plt.subplot(2, 3, 3), plt.imshow(gradient_magnitude, cmap='gray'), plt.title('Gradient Computation')
plt.subplot(2, 3, 4), plt.imshow(edges, cmap='gray'), plt.title('Non-Max Suppression')
plt.subplot(2, 3, 5), plt.imshow(edges, cmap='gray'), plt.title('Hysteresis Thresholding')
plt.subplot(2, 3, 6), plt.imshow(edges, cmap='gray'), plt.title('Canny Edge Detection')
plt.show()
```
